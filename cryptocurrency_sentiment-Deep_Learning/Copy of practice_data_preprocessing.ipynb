```python
crypto_name_e = ['ADA','BTC','BCH','DOGE','EOS','ETH','OMG','SNT','XRP']
crypto_name_k = ['ada','btc','bch','doge','eos','eth','omg','snt','xrp']

for j in crypto_name_e:
  
  X = []

  for i in range(len(DF)):


    if j == DF['coin'][i]:
      a = DF['date'][i]
      b = DF['num'][i]
      c = DF['s'][i]
      d = DF['lang'][i]
      e = DF['coin'][i]

      data = [a,b,c,d,e]
      X.append(data)


  globals()[j] = pd.DataFrame(X)
  globals()[j].columns = ['date','num','s','lang','coin']

#
for j in crypto_name_k:
  
  X = []

  for i in range(len(DF)):


    if DF['coin'][i] == j:
      a = DF['date'][i]
      b = DF['num'][i]
      c = DF['s'][i]
      d = DF['lang'][i]
      e = DF['coin'][i]

      data = [a,b,c,d,e]
      X.append(data)

  globals()[j] = pd.DataFrame(X)
  globals()[j].columns = ['date','num','s','lang','coin']

crypto_e = [ADA,BTC,BCH,DOGE,EOS,ETH,OMG,SNT,XRP]
crypto_k = [ada,btc,bch,doge,eos,eth,omg,snt,xrp]

X = []

for k,l in zip(crypto_e,crypto_k):
  
  for j in range(len(k)):

    for i in range(len(l)):

      if k['date'][j] == l['date'][i]:
        a = k['date'][j]
        b = k['num'][j]
        c = k['s'][j]
        d = l['num'][i]
        e = l['s'][i]
        f = k['coin'][j]

        x = [a,b,c,d,e,f]
        X.append(x)

DF_buble = pd.DataFrame(X)
DF_buble.columns = ['date','num_e','s_e','num_k','s_k','coin']
DF_buble
```
