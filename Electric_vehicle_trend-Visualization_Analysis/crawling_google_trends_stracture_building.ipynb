## 나라이름 가져오기

```python
country_ranking = soup.select( 'div.label-text')
```

```python
country_ranking
```

```python
country_name_list = []
for i in range(len(country_ranking)) :
    a = country_ranking[i]
    b = a.text.strip()
    country_name_list.append(b)
    print(country_name_list)
```

## 각나라 검색율 가져오기

```python
country_ratio = soup.select( 'div.progress-value')
country_ratio
```

```python
country_ratio_check = country_ratio[0]
country_ratio_check
```

```python
country_ratio_check.text.strip()
```

```python
country_ratio_list = []
for i in range(len(country_ratio)) :
    a = country_ratio[i]
    b = a.text.strip()
    country_ratio_list.append(b)
    print(country_ratio_list)
```

## 각 나라 이름과  검색비율 동시에 리스트로 저장하는 동시에 한곳에 저장하기

```python
country_name_ratio_data = []

for i in range(min(len(country_ratio) , len(country_ranking))) :
    a1 = country_ranking[i]
    a2 = country_ratio[i]
    b1 = a1.text.strip()
    b2 = a2.text.strip()
    data = [b1,b2]
    country_name_ratio_data.append(data)
    print(country_name_ratio_data)
```

## 이제 엑셀에다 저장하기 (연습)

```python
df = pd.DataFrame(country_name_ratio_data)
df
```

```python
df = pd.DataFrame(country_name_ratio_data)
df.columns = ['Country Name', 'Search ratio']
df
```

```python
df.to_excel('c://Users//joonc//my_projcets//electric_vehicle_serach_rank.xlsx', index = False)
```
