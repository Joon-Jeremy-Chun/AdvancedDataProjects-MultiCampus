## II.데이터 처리후 모델링

- IQR 3배 기준 밖의 데이터를 outlier로 처리한다
- q25% -(q75-q25)*3
- q75% +(q75-q25)*3

```python
# IQR3 배기준으로 잘라야되는 기준
columns = ['ht','wt','waist','sbp','dbp','fbs','tchol','tg','hdl','ldl','hgb','cr']

for i in columns:
    
    print('=======', i, '=======')
    
    q25, q75 = np.quantile(DF[i], 0.25), np.quantile(DF[i], 0.75)
    IQR = q75 - q25
    
    lower = q25 - IQR*3
    upper = q75 + IQR*3
    
    print('lower :', lower)
    print('upper :', upper, '\n')
```

```python
columns = ['ht','wt','waist','sbp','dbp','fbs','tchol','tg','hdl','ldl','hgb','cr']

for i in columns:
    
    
    q25, q75 = np.quantile(DF[i], 0.25), np.quantile(DF[i], 0.75)
    IQR = q75 - q25
    
    lower = q25 - IQR*3
    upper = q75 + IQR*3
    
    DF_1 = DF.drop(DF.loc[(DF[i]>upper)|(DF[i]<lower)].index)
    print('완료 :', i)


    lowerN = (DF[i]>upper).sum()
    upperN = (DF[i]<lower).sum()
    print('outlier :' , lowerN + upperN)

    
print('outlier가 중복될 수 있음')
DF_1.info()
```
