## 패키지

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools

plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.family'] = 'Malgun Gothic'
```

## 불러오기
- 0번째 제외하고 인덱스 풀기
- 카테고리: 모든 카테고리항목의 문자열을 숫자로 바꾸고 리스트로 가져오기

```python
df = pd.read_csv('c://My_data//17년 세계전기차주제 검색 상대적 추의.csv')
df1=df.iloc[1:]
df1.reset_index()
```

```python
search_ls=list(((np.array(df1['카테고리: 모든 카테고리'].tolist()))))
search_ls1= list(map(int,search_ls))
```

## 달구분되있는 데이터를 각 1월달을 대표로 지정하기
## 년도를 지정하는 과정
- n = 일단 월을 지정한다.(순서대로)
- m = 그다음 매월 1월을 년도의 기준점으로 잡기위해 12로 나누웠을때 나머지가 1인경우만 마크한다
- l = 그 마크에따라 2004년부터 순서대로 기입한다

```python
n = []
a = 0
for i in range(len(search_ls)) :
    a = a + 1
    n.append(a)
```

```python
m = []
for i in n :
    if i % 12 == 1 :
        m.append(1)
    else:
        m.append(0)
```

```python
l = []
a = 2003
for i in m :
    if i == 1 :
        a = a + 1
        l.append(a)
    else :
        l.append('')
```

## 값(list = search_ls) 와 지정한 년도 (list =l) 가지고 그래프 그리기

```python
plt.figure(figsize=(20,7))
plt.ylim((20,100))
plt.xlim([0,213])
plt.title('2004년~ 현재 세계 전기차주제 검색 변화',size =25)
plt.xlabel('기간(년)', labelpad = 10, size =20, weight = 800, )
plt.ylabel('상대적 검색비율', labelpad = 15 , size =20)
y = np.arange(len(search_ls1))
plt.bar(y, search_ls1,color = '#17becf')
plt.xticks(y, l, fontsize=15)


plt.show()
```

## 12개월 평균 이동선 그래프
- 처음 11개월은 앞의 데이터가 없음으로 0으로 간주하고 같이 평균을 낸다.
- m12 이동평균선 리스트 작성
- 그후 그래프

```python
m12 = []
a1=0
a2=0
a3=0
a4=0
a5=0
a6=0
a7=0
a8=0
a9=0
a10=0
a11=0
a12=0
for i in search_ls1 :
    a1 = a2
    a2 = a3
    a3 = a4
    a4 = a5
    a5 = a6
    a6 = a7
    a7 = a8
    a8 = a9
    a9 = a10
    a10 = a11
    a11 = a12
    a12 = i
    
    b = (a12+a11+a10+a9+a8+a7+a6+a5+a4+a3+a3+a2+a1)/12
    m12.append(b)
```

```python
plt.figure(figsize=(20,7))
plt.ylim((20,100))
plt.xlim([0,213])
plt.title('2004년~ 현재 세계 전기차주제 검색 변화(1년 이동평균)',size =25)
plt.xlabel('기간(년)', labelpad = 10, size =20, weight = 800, )
plt.ylabel('상대적 검색비율', labelpad = 15 , size =20)
y = np.arange(len(search_ls1))
plt.plot(y, m12,color = '#17becf')
plt.xticks(y, l, fontsize=15)


plt.show()
```

## 한자리에 합쳐본다.

```python
plt.figure(figsize=(20,7))
plt.ylim((20,100))
plt.xlim([0,213])
plt.title('2004년~ 현재 세계 전기차주제 검색 변화',size =25)
plt.xlabel('기간(년)', labelpad = 10, size =20, weight = 800, )
plt.ylabel('상대적 검색비율', labelpad = 15 , size =20)

y = np.arange(len(search_ls1))
plt.plot(y, m12,color = 'red', linewidth=2.0 ,label = '12개월 평균이동선', linestyle = '--')

plt.bar(y, search_ls1,color = '#17becf')
plt.xticks(y, l, fontsize=15)
plt.yticks(fontsize = 15)
plt.legend(fontsize =14)


plt.show()
```
